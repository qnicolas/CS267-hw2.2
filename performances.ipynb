```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Plot with number of particles vs walltime for 64 threads, 2 Nodes

n = [10000000, 1000000, 100000, 10000, 1000, 100]
mpi_row_parallelism = [50.6815, 3.51696, 0.261632, 0.0576564, 0.0225522, 0.00915077]
serial = [np.nan, 162.000, 6.77681, 0.636754, 0.0552175, 0.00526875]
```

```python
plt.figure(dpi=300)
plt.loglog(n, mpi_row_parallelism, label='MPI')
plt.loglog(n, serial, label='Serial')
# plt.loglog(n, n, linestyle='--')
plt.xlabel('number of particles')
plt.ylabel('wall time (s)')
plt.legend()
plt.show()
```

```python
# Strong scaling

#100k particles
num_tasks = [128, 64, 32, 16, 8, 4, 2 ]
time = [0.226478, 0.364088, 0.422452, 0.783633, 1.45988, 2.91785, 5.67066]

#10k particles
time_10k = [0.0699819, 0.0516664, 0.0688445, 0.0842131, 0.14379, 0.288089, 0.550908]

#1M particles
time_1m = [2.91607, 3.44226, 4.97338,  9.31219, 20.8169, 51.9606, 133.012]

plt.figure(dpi=300)
plt.semilogy(num_tasks, time, label='100K particles')
plt.semilogy(num_tasks, time_10k, label='10K particles')
plt.semilogy(num_tasks, time_1m, label='1M particles')
# plt.loglog(n, n, linestyle='--')
plt.xlabel('number of threads')
plt.ylabel('wall time (s)')
plt.legend()
plt.show()
```

```python
#weak scaling

num_tasks = [128, 64, 32, 16, 8, 4, 2 ]
time_10k_per_thread = [4.02965, 2.37535, 1.43594,  1.29777, 1.17861,  1.15585, 1.13636 ]
time_100k_per_thread = [63.9645, 46.196, 50.6337, 19.063, 15.432, 12.5685, 11.9382]

plt.figure(dpi=300)
plt.semilogy(num_tasks, time_10k_per_thread, label='10K particles/ thread')
plt.semilogy(num_tasks, time_100k_per_thread, label='100K particles/ thread')
# plt.semilogy(num_tasks, time_10k, label='100K particles/ thread')
# plt.loglog(n, n, linestyle='--')
plt.xlabel('number of threads')
plt.ylabel('wall time (s)')
plt.legend()
plt.show()
```

```python
##Compute and communication Times for 100K

num_tasks = [128, 64, 32, 16, 8, 4, 2 ]
communication = [0.100139, 0.097933, 0.068262, 0.057562, 0.067444, 0.052794, 0.046843 ]
compute = [0.237666, 0.315726, 0.441657, 0.770059, 1.49086, 2.91, 5.89549 ]

plt.figure(dpi=300)
plt.plot(num_tasks, communication, label='Communication Time')
plt.plot(num_tasks, compute, label='Compute Time')
plt.xlabel("Number of Tasks")
plt.ylabel("Walltime(s)")

##Compute and communication Times for 1M

num_tasks = [128, 64, 32, 16, 8, 2 ]
communication = [1.900273, 1.451084, 1.47 ,1.141246, 0.771084, 2.460267]
compute = [3.17, 3.72415, 5.85, 8.56, 19, 133 ]

plt.figure(dpi=300)
plt.plot(num_tasks, communication, label='Communication Time')
plt.plot(num_tasks, compute, label='Compute Time')
plt.show()

plt.show()
```

```python
##Compute and communication Times for 1M

num_tasks = [128, 64, 32, 16, 8, 2 ]
communication = [1.900273, 1.451084, 1.47 ,1.141246, 0.771084, 2.460267]
total = [3.17, 3.72415, 5.85, 8.56, 19, 133 ]

plt.figure(dpi=300)
plt.plot(num_tasks[:-1], communication[:-1], label='Communication Time')
plt.plot(num_tasks[:-1], np.asarray(total[:-1]) - np.asarray(communication[:-1]), label='Compute Time')
plt.xlabel("Number of Tasks")
plt.ylabel("Walltime(s)")
plt.legend()
plt.show()
```
